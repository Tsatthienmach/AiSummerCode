### Vẽ đồ thị đường thằng

```
import numpy as np 
import matplotlib.pyplot as plt
```

```
x = np.arange(0,10,0.01)
y = 4*x + 3
```

```
plt.plot(x,y)
plt.show()
```

### Vẽ đồ thị sphere

```
x = np.arange(-10 , 10 , 0.01)
y = x**2
plt.plot(x,y)
plt.show()
```

### Đồ thị hàm cosine

```
x = np.arange(-10, 10 , 0.01)
y = np.cos(x)
plt.plot(x,y)
plt.show()
```

### Đồ thị hàm tanh

```
x = np.arange(-10, 10 , 0.01)
y = np.tanh(x)
plt.plot(x,y)
plt.show()
```
