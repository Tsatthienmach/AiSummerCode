### Vẽ đồ thị đường thằng

```
import numpy as np 
import matplotlib.pyplot as plt
```

```
x = np.arange(0,10,0.01)
y = 4*x + 3
```

```
plt.plot(x,y)
plt.show()
```

### Vẽ đồ thị sphere

```
x = np.arange(-10 , 10 , 0.01)
y = x**2
plt.plot(x,y)
plt.show()
```

### Đồ thị hàm cosine

```
x = np.arange(-10, 10 , 0.01)
y = np.cos(x)
plt.plot(x,y)
plt.show()
```

### Đồ thị hàm tanh

```
x = np.arange(-10, 10 , 0.01)
y = np.tanh(x)
plt.plot(x,y)
plt.show()
```

### Numpy argsort

```
x = np.array([[0 , 5] , [4 , 3] , [2 , 9]])
print(x)
y = np.argsort(x , axis = 0)
print(y)
np.take_along_axis(x ,y , axis = 0 )
```

```
ind = np.unravel_index(np.argsort(x, axis=None), x.shape)
print(ind)
x[ind]  # same as np.sort(x, axis=None)
```
