```
import numpy as np 
import pandas as pd
from numpy import genfromtxt
import matplotlib.pyplot as plt
```

# Linear Regression

```
data_house = genfromtxt('/content/AiSummerCode/week7/my_house_price_prediction.csv',delimiter=',')
X = data_house[:,:-1]
y = data_house[:,-1]
print(type(y))
y = y.reshape(y.shape[0],1)
print(y.shape)
```

```
plt.scatter(X,y)
plt.show()
```

```
def add_bias(X):
  return np.c_[np.ones((X.shape[0],1)),X]
          
def batch_gradient_descent(X,y,n_interations = 100 , learning_rate = 0.01):
  X_b = add_bias(X)
  print("X_b",X_b.shape)
  # Khoi tao tham so theta
  m = X.shape[0]
  theta = np.random.randn(X_b.shape[1],1)
  theta_path = [theta]
  print("Theta",theta.shape)
  losses = []
  for i in range(n_interations):
    # tinh' output
    output = X_b.dot(theta)
    print("output", output.shape)
    # tinh' loss
    loss = (output-y)**2
    print(loss)
    # tinh dao ham cua ham loss
    loss_grd = 2*(output-y)/m
    # tinh dao ham cho cac tham so
    gradients = X_b.T.dot(loss_grd)
    # update
    theta -= gradients*learning_rate
    theta_path.append(theta)
    # mean losss
    mean_loss = sum(loss)/m
    losses.append(mean_loss)
  return losses,theta_path
```

```
n_interations = 50
losses,theta_path = batch_gradient_descent(X,y,n_interations)
plt.scatter(X, y)
data_y = X*theta_path[-1][1]+ theta_path[-1][0]
plt.plot(X,data_y, color="r")
plt.show()
plt.plot(range(n_interations),losses)
```
